```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import polynomial as P

# Read data from the text file
data = np.genfromtxt('hw2.txt', delimiter=',')

# Split data into x and y arrays
x = data[:,1]
y = data[:,0]



# Plot the data and the polynomial fit
plt.plot(x, y, 'o', label='Binding Energy/Nucleon MeV')


# Calculate binding energy of Sn isotopes from 100Sn to 130Sn
binding_energies = []
A_values = []
for A in range(100, 131):
    Z = 50
    N = A - Z
    B = 15.835 * A - 18.33 * A**(2/3) - 23.20 * ((N - Z)**2) / A - 0.714 * (Z**2) / A**(1/3)
    binding_energy_per_nucleon = B / A
    binding_energies.append(binding_energy_per_nucleon)
    A_values.append(A)

# Plot the binding energy per nucleon of Sn isotopes as individual points
plt.plot(A_values, binding_energies, 'o', label='Empirical Mass/Nucleon MeV')

A_values = np.array(A_values, dtype=float)

# Fit the binding energy per nucleon graph as a 5th degree polynomial
coefs = P.polyfit(A_values, binding_energies, 5)
ffit = P.Polynomial(coefs)
plt.plot(A_values, ffit(A_values), label='Fit')

# Add labels and title to the plot
plt.xlabel('Atomic number A')
plt.ylabel('Binding energy/ Empirical Mass per nucleon (MeV)')
plt.title('Comparing Binding Energy and empirical mass of Sn')
plt.legend()

# Save the plot as a .png file
plt.savefig('Hw2q2b.png')

# Show the plot
plt.show()


with open('hw2_q2b.txt', 'w') as f:
    for i in range(len(binding_energies)):
        f.write(f'{binding_energies[i]}, {A_values[i]}\n')
```
